##### A.2) Now use two different methods for tokenization. Compare the results and discuss the differences. Explain them based on how each method works (see manual for methods).

```python
#First, tokenizing with Python using a function. This approach als 'cleans' the text.

def tokenize(text):
    punctuations = '!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~'
    for punctuation in punctuations:
        text = text.replace(punctuation, '')
    text = text.lower() 
    text = text.split()
    return text

tokenized_py = tokenize(text)
print(tokenized_py)
```

```python
#Second, tokenizing with NLTK and Regex. They need to install and import the package.

from nltk.tokenize import RegexpTokenizer

tokenizer = RegexpTokenizer('\w+')
tokenized_nltk = tokenizer.tokenize(text.lower())
print(tokenized_nltk)
```

```python
#Third, tokenizing with Spacy. They need to install and import the package.

import spacy

nlp = spacy.load("en_core_web_sm")
processed_text = nlp(text.lower())

tokenized_spacy = [token.text for token in processed_text if not token.is_punct]
print(tokenized_spacy)
```

##### B.6) Convert the words in the tweets to lowercase. Create a new column for that to keep the original tweets.

```python
#code
df["tweet2"] = df["tweet"].str.lower()

#I usually check if everything went as planned
df.head(5)
```

##### B.7) Remove the numbers from the tweets.

```python
#They will need to work with string.digits for this operation.
import string

df['tweet2'] = df['tweet2'].str.translate(str.maketrans(' ', ' ', string.digits))

```

##### B.8) Remove punctuation from the tweets.

```python
#They will ned to work with string.punctuation
df['tweet2'] = df['tweet2'].str.translate(str.maketrans(' ', ' ', string.punctuation))
```

##### B.9) Strip excessive white spaces from the tweets.

```python
#Some info on .strip() https://www.simplilearn.com/tutorials/python-tutorial/strip-in-python
df['tweet2'] = df['tweet2'].str.strip()
```

##### B.10) Remove the stopwords from the tweets. Compare your "cleaned-up" tweets with the originals.

```python
#We're using NLTK's stopword list for English. Later on, we will discuss that stopword list can be expanded by adding terms.
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = stopwords.words('english')

df['tweet2'] = df['tweet2'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
```

##### B.15) Bonus: try simple topic modeling with the hashtags using a for-loop and LdaModel. Can you discern distinct topics?

```python
#For-loop to retrieve the tags 
#Students can then discuss if there are indeed distinct topics and how they would label them
#We will discuss LDA topic modeling in week 2 and address issues of pre-processing, determing k, validation etc.

tags = [[tag.lower() 
         for tag in re.findall("#\w+", tweet)] 
        for tweet in df["tweet"]]
voca = corpora.Dictionary(tags)
corpus = [voca.doc2bow(doc) for doc in tags]
m = models.LdaModel(corpus, num_topics=10,
  id2word=voca,
  distributed=False, random_state=123)
m.print_topics() 
```
